### Multiple Logging Logging

Can Create multiple loggers for different parts of your application.

```python
import logging

## create a logger for module1
logger1 = logging.getLogger("module1")        # getLogger :- To create a logger for a specific module.
logger1.setLevel(logging.DEBUG)

## create a logger for module1
logger2 = logging.getLogger("module2")        
logger2.setLevel(logging.WARNING)

## configuring logging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
    )
```

```python
## log messages with different loggers

logger1.debug("This is a debug message for mudule1")
logger2.warning("This is a warning messages for module2")
logger2.error("This is a error message")
```
