### Python Logging


when we do an end to and project we have to probably log each and every thing that is happening with respect to the execution.Whether its a warning, whether its an error everything needs to be logged.

SO that whenever we debug, the entire application will be able to see what king of errors, if any, error is occuring in the application, we'll be able to understand it.

```python
import logging

## Configure the basic loggign settings
logging.basicConfig(level=logging.DEBUG)     # Since developing appliction we use debug

# log messages with different severity levels
logging.debug("This is a debug message")
logging.info("This is an info messages")
logging.warning("This is a warning message")
logging.error("This is an error messages")
logging.critical("This is a critical messages")

```

```python
import logging

# Reset all previous logging configs
for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

### Log Levels

Default levels are:

- DEBUG : Detailed information, typically of interest only when diagnosing problems.
- INFO: Confirmation that things are working as expected.
- WARNING: An indication that something unexpected happened or indicative of some problem in the near future (e.g., ‘disk space low’). The software is still working as expected.
- ERROR: Due to a more serious problem, the software has not been able to perform some function.
- CRITICAL: A very serious error, indicating that the program itself may be unable to continue running.
