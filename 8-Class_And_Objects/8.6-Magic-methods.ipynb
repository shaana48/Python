### Magic Methods

Those are predefined methods in python that you can override to change the behaviour of your objects.
Some common magic methods are:

```python
'''
__int__ : Initializes a new instance of a class
__str__ : Returns string representation of an object
__repr__ : Returns an official string representation of an object
__len__ : Returns the length of an object
__getitem__ : Gets an item from a contianer
__setitem__ : Sets an item in a container
'''
```

```python
class Person:
    pass

person=Person()
dir(person)
```

```python
## Bsic  Magic methods

class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

person=Person("krish",34)
print(person)
```

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"{self.name},{self.age} years old"
    
    def __repr__(self):
        return f"Person(name={self.name},age={self.age})"
    
person=Person("krish",34)
print(person)
print(repr(person))
```
