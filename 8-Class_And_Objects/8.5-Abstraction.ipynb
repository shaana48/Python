## Abstraction

This method is like giivng you the necessary feature of an object  and just hiding the complex implementation detail 

```python
from abc import ABC,abstractmethod

## Abstract class
class Vehicle(ABC):
    def drive(self): 
        print("The vehicle is used foer driving")

    @abstractmethod
    def start_engine(self):
        pass

## child class
class Car(Vehicle):
    def start_engine(self):
        print("Car enginer started")

def operate_vehicle(vehicle):
    vehicle.start_engine()

car=Car()
operate_vehicle(car)        
```
