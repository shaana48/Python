##### What are EXCEPTIONS?
Exceptions are events that disrupt the normal flow of a program.They occur when an error is encountered during program excecution.Common exceptions include:

ZeroDivisionError : Dividing by zero
FileNotFountError : File not found
ValueError : Invalid Value
TypeError : Invalid type

##### Try,Except,Else

```python
try:
    num=int(input("Enter a number:"))
    result=10/num
except ValueError:
        print("This is not a valid number")
except ZeroDivisionError:
      print("You can't divide by zero")
except Exception as ex:
      print(ex)      
else:
      print(f"The result is {result}")
                 
```

##### Try , Except , Else and Finally

```python
try:
    num=int(input("Enter a number:"))
    result=10/num
except ValueError:
        print("This is not a valid number")
except ZeroDivisionError:
      print("You can't divide by zero")
except Exception as ex:
      print(ex)      
else:
      print(f"The result is {result}")
finally:
      print("Excecution complete.")       # MUst exceute whether there is an erroe or no error
```
