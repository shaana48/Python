```python
## while loop
## The while loop continues to execute as long as the condition is True

count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

```python
## loop Control Statements

# break
#The break statement exits the loop permenantly 

for i in range(10):
    if i==5:
        break
    print(i)

```

```python
## continue
## The continue statement skips the current iteration and continues with the next

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass
## the pass statement is anull operation ie: it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loop
## A loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples
# 1- calculate the sum of first N natural numbers using a while & for loop

# WHILE

n=int(input("Enter number"))
sum=0
count=1
while count<=n:
    sum=sum+count
    count=count+1
print(f"Sum of first {n} natural number is {sum}")


```

```python
## FOR

n=int(input("Enter number"))
sum=0

for i in range(1,n+1):
    sum=sum+i
print(f"Sum of first {n} natural number is {sum}")    
```

```python
# 1- prime numbers btwn 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)    
```
