**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
yr=int(input("Enter a year"))
if yr % 400 == 0:
    print(f"{yr} is a leap year")
elif yr % 4 == 0 and yr % 100 != 0:
    print(f"{yr} is a leap year")
else:
    print(f"{yr} is not a leap year")     
```
