2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
dtype = [('x',int),('y',int)]
data = [(1,2),(3,4),(5,6)]
points = np.array(data, dtype=dtype)
print("Points:\n", points)

# Euclidean distance
x = points['x']
y = points['y']

dx = x[:, np.newaxis] - x[np.newaxis, :]
dy = y[:, np.newaxis] - x[np.newaxis, :]

# 1- [:, np.newaxis] :- Convert 1D(3) to 2D Column(3,1)
#                    [1,3,5] :- [[1],
#                                [3],
#                                [5]]
# 2- [np.newaxis, :] :- Converts it to 2D row (1,3)
#                    [[1],
#                     [3], :- [[1,4,8]]
#                     [5]]
# 1-2 :- oru matrix kittum 1-2 cheythath...dx l x nte differences m dy l y nte differences m aayitlla oru matrix 

distance = np.sqrt(dx**2 + dy**2)
print("Euclidean distances:")
print(distance)

# dx**2 → squares all x-differences
# dy**2 → squares all y-differences
# dx**2 + dy**2 → sums the squares
# np.sqrt(...) → takes the square root → Euclidean distance
```

### Assignment 10: Masked Arrays

1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

```python
array = np.random.randint(1,17, size=(4,4))
print(f"Array :\n", array)

masked_array = np.ma.masked_greater(array, 10)         # masks all elements >10......Masked elements are ignored in computations
print(f"Masked Array :\n", masked_array)

unmasked_sum = np.sum(masked_array)
print(f"Sum of Unmasked Array :\n", unmasked_sum)


```

2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
array = np.random.randint(1,10, size=(3,3))
print(f"Array :\n", array)

mask = np.eye(3, dtype=bool)
# np.eye(3) → creates a 3×3 identity matrix
# dtype=bool → converts 1 → True, 0 → False   :- Purpose: True marks diagonal elements → these will be masked                    
masked_array = np.ma.array(array, mask=mask)
print(f"Masked Array :\n", masked_array)

mean = np.mean(masked_array)
print(f"Mean of Unmasked elements :\n", mean)

masked_array[mask] = mean
print(f"Required Array :\n", masked_array)
```
