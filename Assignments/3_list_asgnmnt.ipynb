```python
## Assignment 10: Flattening a Nested List
# Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

def flattend(list1):
    print(f"original list is {list1}")
    flat=[]
    for i in list1:
        for j in i:
            flat.append(j)
    print(f"flattened list is {flat}")
        
flattend([[1,2,3],[4,5,6],[7,8,9]])       
```

```python
## Assignment 13: List Reversal
# Write a function that takes a list and returns a new list with the elements in reverse order. 
# Print the original and reversed lists.

def revers(list1):
    print("Original list is ", list1)
    list1.reverse()
    return f"reversed list is {list1}"

revers([1,2,3,4,5])
```

```python
## Assignment 14: List Rotation
# Write a function that rotates a list by n positions. Print the original and rotated lists.

def rotation(list, n):
    new=list[n:]+list[:n]
    return new
rotation([2,4,6,8,10], 3)
```

```python
## Assignment 15: List Intersection
#Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. 
# Print the intersected list.

def intersection(lst1,lst2):
    new=[]
    for i in lst1:
       new.append(i)
    for j in lst2:
       if j not in new:
        new.append(j)
    return new
intersection(lst1 = [2,4,6,8,10],lst2 = [3, 4, 5, 6, 7])



         


```
