```python
## Tuple Methods
print(numbers.count(1))      # How many time the element presented
print(numbers.index(3))      # Gives index of that value
```

```python
## Packing a tuple

packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking a tuple

a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested List
lst=[[1,2,3,4],[6,7,8,9],(1,"hello",3.14,"c")]
lst[0]
lst[0][2]
lst[2][0:3]
```

```python
## Nested Tuple
nested_tuple=((1,2,3),("a","b","c"),(True,False))

## Acces elements inside a tuple

print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
```
