### SETS

#### Built-in data types used to store collection of unique items
#### Do not allow Duplicate elements

```python
## Create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_empty_set=set([1,2,3,4,5])
print(my_empty_set)
```

```python
my_empty_set=set([1,2,3,6,5,4,5,6])
print(my_empty_set) 
```

#### Basic Set Operations

```python
## Adding an element

my_set.add(7)
print(my_set)
```

```python
## Removing an element
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

##### Discard : Remove an element from a set if its a memeber
#####           Unlike remove(),the discard() doesnt raise error

```python
my_set.discard(10)
print(my_set)
```

```python
## pop method

removed_element=my_set.pop()    # pop removes 1 st elemnt
print(removed_element)
print(my_set)
```

```python
## clear all elements
my_set.clear()
print(my_set)
```

```python
## Set Membership Test : Whether an element is a member of a set

my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
```

```python
## UNION

union_set=set1.union(set2)
print(union_set)
```

```python
## INTERSECTION

intersection_set=set1.intersection(set2)
print(intersection_set)
```

```python
## intersection update

set1.intersection_update(set2)
print(set1)
```

#### intersection() → Creates a new set with the common elements. ✅ The original sets stay the same.

#### intersection_update() → Changes the first set itself and removes all elements that are not common. ❌ No new set is made.

```python
## Counting Unique words in text

text="In this tutorial we are discussing abt sets"
words=text.split()

## convert list of words to set to get unique words
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```

```python
set1 = set(range(1, 6))
set2 = set(range(2, 11, 2))
print(f"Union: {set1 | set2}")
print(f"Intersection: {set1 & set2}")
print(f"Difference (set1 - set2): {set1 - set2}")
print(f"Symmetric Difference: {set1 ^ set2}")
```
