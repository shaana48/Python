```python
# why functions?

num=28
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd """
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")  
```

```python
## Calling the Function

even_or_odd(24)
even_or_odd(25)
```

```python
# Function with multiple parameter

def add(a,b):                   # def add(a,b):
    return a+b                  #      return a+b  
                             
result=add(2,4)                 # add(2,4)
print(result)
```

```python
# Default parameter

def greet(name):
    print(f"Hello {name} Welcome to the paradise")

greet("Shana")    
```

```python

def greet(name):
    print(f"Hello {name} Welcome to the paradise")

greet()
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome to the paradise")

greet()
greet("shana")
```

```python
# Variable Length Argument
# Positional & Keyword Argument

def print_numbers(*shana):             
     for number in shana:
          print(number)   
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Shana")
```

```python
# Positional Arguments

def print_numbers(*args):             
     for number in args:
          print(number) 
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Shana")
```

```python
# Keyword Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Shana",age=23,country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
          print("Positional argument :",val) 
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Fthma",name="Shana",age=23,country="India")
```

```python
# Return multiple parameters

def multiply(a,b):
    return a*b,a

multiply(2,3)
```
