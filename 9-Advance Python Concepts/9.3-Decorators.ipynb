#### DECORATORS

Decorators are powrful and flexible feature in python that allows you to modify the behaviour of a function or class method.They are commonly used to add functionality to functions or methods without modifying their actual code.

```python
## function copy
## closures
## decorators
```

```python
## function copy

def welcome():
    return "Welcome to the advanced python course"

welcome()
```

```python

wel=welcome
print(wel())
del welcome
print(wel())
```

```python
## Closures : Basically means a function inside a function

def main_welcome(msg):
    
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome("Welcome Everyone")
```

```python
### Decorator

def main_welcome(func):
    
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
def course_intro():
    print("This is an advanced python course")

course_intro()    
```

```python
main_welcome(course_intro)
```

```python
@main_welcome
def course_intro():
    print("This is an advanced python course")
```

WE DIDNT CALL THE ENTIRE FUNCTION THIS MEANS WHENEVER WE DO THIS THIS GO INTO THAT PARTICULAR FUNCTIN   THIS IS CALLED DECORATOR.

```python
## Decorator

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called")
        func()
        print("Something is happening after the function is called")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## Decorators with Arguments

def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator        
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```
