```python
## Data Serialization

import json
data={'name':"Shana",'age':24}

json_str=json.dumps(data)            ## Converting
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)       ## Converting back to dictionary
print(parsed_data)
print(type(parsed_data))
```

```python
## csv

import csv
with open('example.csv',mode='w',newline='')as file:      # Craeted a file called example.csv with contents
    writer=csv.writer(file)                          
    writer.writerow(['name','age'])
    writer.writerow(['Shana',24])
    

with open('example.csv',mode="r") as file:                # Readed that file
    reader=csv.reader(file)
    for row in reader:
        print(row)
        
```

```python
## datatime

from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
## time

import time
print(time.time())
time.sleep(2)
print(time.time())
```
