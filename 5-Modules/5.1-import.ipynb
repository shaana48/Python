#### Importing MODULES in Python : Modules & Packages 

In python, modules and packages help organize and reuse code.
Here's a comprehensive guide on how to import them

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math import *
print(sqrt(16))
print(pi)
```

** means importing all the functions that are available in that module.   You dont have to specify

```python
from package.maths import addition
addition(2,4)
```

```python
from package import maths
maths.addition(2,4)
```
