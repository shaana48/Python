##### FILE OPERATION - Read & Write Files
File handling is a crucial part of any programming language.Python provides built-in functions and methods to read from and write to files,both text and binary.

```python
### Read a Whole File

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## REAd a file line by line

with open ('Example.txt','r') as file:
    for line in file:
        print(line)
```

```python

with open ('Example.txt','r') as file:
    for line in file:
        print(line.strip())           # strip() removes the new line character
```

```python
## Writing a file(Overwriting)

with open('Example.txt','w') as file:
    file.write('Hello World!\n')
    file.write('this is a new line')
```

```python
## Write a file (without OVERWRITING) 

with open ('Example.txt','a') as file:
    file.write("Append  operation taking place\n!")
```

```python
##Writng list of lines to a file
lines=['First line \n','Second line\n','Third line\n']
with open('Example.txt','a')as file:
    file.writelines(lines)
```

```python
## Writing a Binary Files

data=b'\x00\x01\x02\x03\x04'
with open('Example.bin','wb') as file:
    file.write(data)
```

```python
# Reading a Binay file

with open('Example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
# Read the content from a source file and write to a destination text file
# Copying text file

with open('Example.txt','r') as source_file:
    content=source_file.read()
    with open('Destination.txt','w') as Destination_file:
        Destination_file.write(content)
```
