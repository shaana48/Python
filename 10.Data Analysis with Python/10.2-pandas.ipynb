#### PANDAS-DataFrame And Series

Pandas is a powerful data manipulation library in python,widely used for data analysis and data cleaning.it provides two primary data structures:Series and DataFrame.A Series is a one-dimensional array-ike object,while a DataFrame is a two dimensional ,size mutable and potentially heterogeneous tabular data structure with labeled axes(rows and columns).

```python
##  Series 
# A Pandas series is a one dimensional array-like object that can hold any data type.It is similar to acolumn in a table.

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)

```

```python
## Create a series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30,]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
## Create a DataFrame from a dictionary of lists

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,26],
    'City':['Banglore','New york','Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a dataframe from a list of dictionaries

data=[
    {'Name':'Krish','Age':24,'City':'Banglore'},
    {'Name':'John','Age':24,'City':'Banglore'},
    {'Name':'Jack','Age':24,'City':'Banglore'},
    {'Name':'Peter','Age':24,'City':'Banglore'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
